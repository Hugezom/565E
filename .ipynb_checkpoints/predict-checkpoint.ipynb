```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import scipy.stats as ss
```

```python
df = pd.read_csv("cleaned.csv")
```

```python
data = df[df.symbol == "BTC/USDT"]
data.loc[:, "transferred_avg_value"] = data.transferred_value/data.transferred_number
data = data.drop(data.index[data.transferred_avg_value.isna()])

dta = data.sort_values(["block_number", "gas_price"],
                      ascending=False).drop_duplicates("block_number", keep="first")
dta = dta.set_index("block_number")
data = data.set_index("block_number")
```

```python
dta = dta.drop(dta.index[dta.gas_price>=dta.gas_price.mean() + 2*dta.gas_price.std()])
```

```python
plt.figure(figsize=(9, 6), dpi=300)
dta.gas_price.plot()
```

```python
def evaluate(factor):
    corr = np.corrcoef(factor, dta.gas_price)[0, 1]
    return corr

def mean(array, N=None):
    if isinstance(N, int):
        array = array[:N]
    return sum(array) / len(array)


def rsi(array, N):
    if isinstance(N, int):
        array = array[:N]
    su = 0
    sd = 0
    for value in array:
        if value < 0:
            sd -= value
        else:
            su += value

    value = su + sd
    value = (su - sd) / value if value != 0 else 0
    return value
```

# Factor

```python
f0 = []
for bn in dta.index:
    tmpv = data.loc[bn-1:bn-2, "gas_price"].mean()
    if np.isnan(tmpv):
        tmpv = 0
    f0.append(tmpv)

f1 = []
for bn in dta.index:
    tmpv = data.loc[bn-1:bn-100, "gas_price"].mean()
    if np.isnan(tmpv):
        tmpv = 0
    f1.append(tmpv)

f2 = []
for bn in dta.index:
    tmpv = data.loc[bn-1:bn-100, "gas_price"].std()
    if np.isnan(tmpv):
        tmpv = 0
    f2.append(tmpv)

f3 = []
for bn in dta.index:
    values = []
    for i in range(1, 100+1):
        lbn = bn-i
        if lbn in dta.index:
            tmpv = dta.loc[lbn, "gas_price"]
            values.append(tmpv)
        else:
            values.append(0)
    f3.append(np.argmax(values))

f4 = []
for bn in dta.index:
    values = []
    for i in range(1, 4+1):
        lbn = bn-i
        if lbn in dta.index:
            tmpv = dta.loc[lbn, "gas_price"]
            values.append(tmpv)
        else:
            values.append(0)
    f4.append(np.argmin(values))
    
f5 = []
for bn in dta.index:
    values = []
    for i in range(1, 10+1):
        lbn = bn-i
        if lbn in dta.index:
            tmpv = dta.loc[lbn, "transferred_avg_value"]
            values.append(tmpv)
        else:
            values.append(0)
    f5.append(mean(values, 2))
    
f6 = []
for bn in dta.index:
    tmpv1 = np.nan_to_num(data.loc[bn-1:bn-1, "gas_price"].count())
    tmpv2 = np.nan_to_num(data.loc[bn-1:bn-100, "gas_price"].count())
    f6.append(tmpv1-tmpv2)
    
f7 = []
for bn in dta.index:
    tmpv = np.nan_to_num(data.loc[bn-1:bn-100, "can_abi"].sum()/data.loc[bn-1:bn-100, "can_abi"].sum())
    f7.append(tmpv)
```

```python
def _fold(array, t):
    folded_array = np.lib.stride_tricks.as_strided(
        array,
        shape=(len(array) - t + 1, t),
        strides=(array.itemsize * 1, array.itemsize * 1),
    )
    return folded_array
```

# Training

```python
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error as mae
import random
```

```python
Y = dta.gas_price.values
X = np.c_[f0, f1, f2, f3, f4, f5, f6, f7]

_id = np.arange(dta.shape[0])
random.seed(10)
random.shuffle(_id)

Y = Y[_id]
X = X[_id]
```

```python
kf = KFold(n_splits=5)
result = []
cost = []
pos = []
pri_pct = []

for train, test in kf.split(X):
    X_train = X[train]
    X_test = X[test]
    
    Y_train = Y[train]
    Y_test = Y[test]
    
    model = sm.OLS(Y_train, sm.add_constant(X_train)).fit()
    
    y_pred = model.predict(sm.add_constant(X_test))
    result.append(mae(y_pred, Y_test))
    
    cost.append(sum(y_pred)/len(test))
    _pos = []
    for i in range(len(test)):
        _pos.append((data.loc[[dta.iloc[_id[test[i]]].name], "gas_price"]>y_pred[i]).sum()+1)
    _pos = np.array(_pos)
    pos.append(np.mean(_pos))
    pri_pct.append((_pos==1).sum()/len(_pos))
    
print(np.mean(result))
print(np.mean(cost))
print(np.mean(pos))
print(np.mean(pri_pct))
```
